# 01_cleaning

Responsabilite:
- nettoyer le classeur brut `data/consolidation_maritime.xlsx`
- produire la base propre commune au projet maritime

Entrees:
- `data/consolidation_maritime.xlsx`

Sorties:
- `data/maritime_clean.csv`

Execution:
- ouvrir le notebook depuis la racine du projet
- executer toutes les cellules après avoir selectionne l'environnement Python


```python
from pathlib import Path
import sys

project_root = None
cwd = Path.cwd().resolve()
for candidate in (cwd, *cwd.parents):
    if (candidate / "src").exists() and (candidate / "data").exists():
        project_root = candidate
        break

if project_root is None:
    raise FileNotFoundError(f"Impossible de trouver la racine du projet depuis {cwd}")

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

import pandas as pd
from src.paths import get_data_dir
from src.cleaning_utils import load_raw_data, clean_raw_data, reorganize_columns

data_dir = get_data_dir()
source = data_dir / "consolidation_maritime.xlsx"

# Chargement
df = load_raw_data(source)

# Nettoyage
df = clean_raw_data(df)

# Réorganisation
df = reorganize_columns(df)

# Export
df.to_csv(data_dir / "maritime_clean.csv", index=True)
```
